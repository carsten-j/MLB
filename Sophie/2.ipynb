```python
import matplotlib.pyplot as plt
import numpy as np


# Define the function f(x) = -0.0034x^3 + 0.077x^2 - 0.3x + 2.46
def f(x):
    return -0.0034 * x**3 + 0.077 * x**2 - 0.3 * x + 2.46


def df(x):
    return -3 * 0.0034 * x**2 + 2 * 0.077 * x - 0.3
```

```python
# Create x values for plotting
x = np.linspace(-10, 30, 1000)
y = f(x)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, "b-", linewidth=2, label="f(x) = -0.0034x³ + 0.077x² - 0.3x + 2.46")

# Fill the area between y=0 and f(x) from x=0 to x=14.5
x_fill = np.linspace(0, 14.5, 500)
y_fill = f(x_fill)
plt.fill_between(
    x_fill, 0, y_fill, alpha=0.3, color="blue", label="Area from x=0 to x=14.5"
)

plt.grid(True, alpha=0.3)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of f(x) = -0.0034x³ + 0.077x² - 0.3x + 2.46")
plt.legend()
plt.axhline(y=0, color="k", linestyle="-", alpha=0.3)
plt.axvline(x=0, color="k", linestyle="-", alpha=0.3)
plt.axvline(x=14.5, color="r", linestyle="--", alpha=0.7, label="x = 14.5")
plt.show()
```

```python
from scipy import integrate

# Compute the integral of f(x) from 0 to 14.5
integral_result, error = integrate.quad(f, 0, 14.5)

print(f"Integral of f(x) from 0 to 14.5: {integral_result:.6f}")
print(f"Estimated error: {error:.2e}")

# Also compute using analytical integration
# f(x) = -0.0034x³ + 0.077x² - 0.3x + 2.46
# F(x) = -0.0034x⁴/4 + 0.077x³/3 - 0.3x²/2 + 2.46x
# F(x) = -0.00085x⁴ + 0.025667x³ - 0.15x² + 2.46x


def F(x):
    """Antiderivative of f(x)"""
    return -0.00085 * x**4 + (0.077 / 3) * x**3 - 0.15 * x**2 + 2.46 * x


# Analytical result using fundamental theorem of calculus
analytical_result = F(14.5) - F(0)
print(f"\nAnalytical integral: {analytical_result:.6f}")
print(f"Difference between methods: {abs(integral_result - analytical_result):.2e}")
```

```python
import numpy as np


def dh(x):
    """
    Derivative of h(x) = (x+1)*exp(x^2+x)
    Using product rule: (fg)' = f'g + fg'
    where f = (x+1) and g = exp(x^2+x)

    f'(x) = 1
    g'(x) = exp(x^2+x) * (2x+1)  [chain rule]

    Therefore: dh(x) = 1*exp(x^2+x) + (x+1)*exp(x^2+x)*(2x+1)
                     = exp(x^2+x) * [1 + (x+1)*(2x+1)]
                     = exp(x^2+x) * [1 + (2x^2 + 3x + 1)]
                     = exp(x^2+x) * (2x^2 + 3x + 2)
    """
    return np.exp(x**2 + x) * (2 * x**2 + 3 * x + 2)


# Test the derivative
x_test = np.array([0, 1, -1, 2])
print("Testing dh(x) at various points:")
for x_val in x_test:
    print(f"dh({x_val}) = {dh(x_val):.6f}")

# Also show the original function values for comparison
print("\nOriginal function h(x) values:")
for x_val in x_test:
    h_val = (x_val + 1) * np.exp(x_val**2 + x_val)
    print(f"h({x_val}) = {h_val:.6f}")
```

```python
# Solve dh(x) = 0
# Since dh(x) = exp(x^2+x) * (2x^2 + 3x + 2)
# and exp(x^2+x) > 0 for all x, we need to solve:
# 2x^2 + 3x + 2 = 0

# Using the quadratic formula: x = (-b ± √(b²-4ac)) / 2a
# where a = 2, b = 3, c = 2

a = 2
b = 3
c = 2

# Calculate discriminant
discriminant = b**2 - 4 * a * c
print(f"Quadratic equation: {a}x² + {b}x + {c} = 0")
print(f"Discriminant = b² - 4ac = {b}² - 4({a})({c}) = {discriminant}")

if discriminant > 0:
    x1 = (-b + np.sqrt(discriminant)) / (2 * a)
    x2 = (-b - np.sqrt(discriminant)) / (2 * a)
    print("\nTwo real solutions:")
    print(f"x₁ = {x1:.6f}")
    print(f"x₂ = {x2:.6f}")

    # Verify the solutions
    print("\nVerification:")
    print(f"dh({x1:.6f}) = {dh(x1):.10f}")
    print(f"dh({x2:.6f}) = {dh(x2):.10f}")

elif discriminant == 0:
    x = -b / (2 * a)
    print("\nOne repeated real solution:")
    print(f"x = {x:.6f}")
    print(f"Verification: dh({x:.6f}) = {dh(x):.10f}")

else:
    real_part = -b / (2 * a)
    imag_part = np.sqrt(-discriminant) / (2 * a)
    print("\nNo real solutions. Complex solutions:")
    print(f"x₁ = {real_part:.6f} + {imag_part:.6f}i")
    print(f"x₂ = {real_part:.6f} - {imag_part:.6f}i")
    print("\nSince discriminant < 0, there are no real values of x where dh(x) = 0")
```
