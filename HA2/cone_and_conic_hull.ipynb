```python
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with 3 subplots for different visualizations
fig = plt.figure(figsize=(15, 12))

# Example 1: 2D cone (a sector in 2D)
ax1 = fig.add_subplot(221)
theta = np.linspace(np.pi / 6, np.pi / 3, 100)
r = np.linspace(0, 5, 20)
theta_grid, r_grid = np.meshgrid(theta, r)
x = r_grid * np.cos(theta_grid)
y = r_grid * np.sin(theta_grid)
ax1.scatter(x, y, s=5, color="blue", alpha=0.6)
ax1.scatter(0, 0, s=50, color="red")  # Origin
ax1.set_xlim(-5, 5)
ax1.set_ylim(-1, 5)
ax1.set_aspect("equal")
ax1.grid(True)
ax1.set_title("2D Cone (Sector)")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

# Example 2: 3D cone
ax2 = fig.add_subplot(222, projection="3d")
theta = np.linspace(0, 2 * np.pi, 30)
r = np.linspace(0, 5, 10)
t_grid, r_grid = np.meshgrid(theta, r)
z = r_grid
x = r_grid * np.cos(t_grid) * 0.5
y = r_grid * np.sin(t_grid) * 0.5
ax2.scatter(x, y, z, c="blue", alpha=0.6)
ax2.scatter(0, 0, 0, s=50, color="red")  # Origin
ax2.set_title("3D Cone")
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_zlabel("z")

# Example 3: Conic Combination of two vectors
ax3 = fig.add_subplot(223)
v1 = np.array([2, 1])
v2 = np.array([0, 3])

# Plot the vectors
ax3.arrow(
    0,
    0,
    v1[0],
    v1[1],
    head_width=0.2,
    head_length=0.3,
    fc="blue",
    ec="blue",
    label="v1",
)
ax3.arrow(
    0,
    0,
    v2[0],
    v2[1],
    head_width=0.2,
    head_length=0.3,
    fc="green",
    ec="green",
    label="v2",
)

# Plot some conic combinations
lambda1_values = [0.5, 1, 1.5, 2]
lambda2_values = [0.3, 0.7, 1, 1.3]

for l1 in lambda1_values:
    for l2 in lambda2_values:
        comb = l1 * v1 + l2 * v2
        ax3.plot([0, comb[0]], [0, comb[1]], "r-", alpha=0.3)
        ax3.plot(comb[0], comb[1], "ro", alpha=0.5)

ax3.grid(True)
ax3.set_xlim(-1, 6)
ax3.set_ylim(-1, 6)
ax3.set_aspect("equal")
ax3.set_title("Conic Combinations of Two Vectors")
ax3.set_xlabel("x")
ax3.set_ylabel("y")
ax3.legend()

# Example 4: Conic Hull
ax4 = fig.add_subplot(224)

# Generate some random points
np.random.seed(42)
points = np.random.rand(10, 2) * 4
points = np.vstack([points, [0, 0]])  # Add origin for visualization

# Plot the points
ax4.scatter(points[:, 0], points[:, 1], color="blue", s=50, label="Original points")

# Plot rays from origin through each point
for i in range(len(points)):
    if not np.array_equal(points[i], [0, 0]):  # Skip origin
        ray_end = points[i] * 2  # Extend the ray
        ax4.plot([0, ray_end[0]], [0, ray_end[1]], "r-", alpha=0.3)

# Fill the conic hull with light color
hull_theta = np.linspace(0, 2 * np.pi, 100)
hull_r = 5
hull_x = hull_r * np.cos(hull_theta)
hull_y = hull_r * np.sin(hull_theta)
ax4.fill(hull_x, hull_y, color="yellow", alpha=0.2)

ax4.grid(True)
ax4.set_xlim(-5, 5)
ax4.set_ylim(-5, 5)
ax4.set_aspect("equal")
ax4.set_title("Conic Hull Illustration")
ax4.set_xlabel("x")
ax4.set_ylabel("y")

plt.tight_layout()
plt.show()
```
