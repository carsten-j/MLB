```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters
mu = 1
learning_rates = [0.0001, 0.001, 0.01]
colors = ["red", "blue", "purple"]
labels = [" = 0.0001", " = 0.001", " = 0.01"]

# Create array of iterations
t = np.arange(0, 1201)

# Create figure with two subplots (log scale and linear scale)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))

# Plot with log scale
for i, lr in enumerate(learning_rates):
    convergence_factor = (1 - lr * mu) ** t
    ax1.plot(t, convergence_factor, color=colors[i], label=labels[i], linewidth=2)

ax1.set_title(
    "Convergence Factor $(1-)^t$ for Different Learning Rates (Log Scale)",
    fontsize=14,
)
ax1.set_xlabel("Iteration (t)", fontsize=12)
ax1.set_ylabel("Convergence Factor $(1-)^t$", fontsize=12)
ax1.legend(fontsize=12)
ax1.grid(True)
ax1.set_yscale("log")  # Using log scale
ax1.set_ylim(1e-6, 1.1)

# Plot with linear scale
for i, lr in enumerate(learning_rates):
    convergence_factor = (1 - lr * mu) ** t
    ax2.plot(t, convergence_factor, color=colors[i], label=labels[i], linewidth=2)

ax2.set_title(
    "Convergence Factor $(1-)^t$ for Different Learning Rates (Linear Scale)",
    fontsize=14,
)
ax2.set_xlabel("Iteration (t)", fontsize=12)
ax2.set_ylabel("Convergence Factor $(1-)^t$", fontsize=12)
ax2.legend(fontsize=12)
ax2.grid(True)
ax2.set_ylim(0, 1.1)

plt.tight_layout()
plt.savefig("convergence_factor_plot.png", dpi=300)
plt.show()

# Print specific values at certain iterations
iterations = [100, 200, 500, 1000, 1200]
print("Convergence factor (1-)^t values:")
print(f"{'Iteration':<10} {' = 0.0001':<20} {' = 0.001':<20} {' = 0.01':<20}")
print("-" * 70)

for it in iterations:
    values = [(1 - lr * mu) ** it for lr in learning_rates]
    print(f"{it:<10} {values[0]:<20.10e} {values[1]:<20.10e} {values[2]:<20.10e}")
```
