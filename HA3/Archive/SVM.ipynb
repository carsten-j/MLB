```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.metrics import zero_one_loss
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

plt.style.use("ggplot")
```

```python
# ------------------------------------------------------------------
# 1. Load data
# ------------------------------------------------------------------
df = pd.read_csv("nonlinear_svm_data.csv")  # x, y, label
X = df[["x", "y"]].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

y = df["label"].values
y = 2 * y - 1  # Transforms 0->-1 and 1->+1

n_obs = X.shape[0]
print(f"Number of observations: {n_obs}")
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
x_min, x_max, y_min, y_max = -1, 1, -1, 1
ax.set(xlim=(x_min, x_max), ylim=(y_min, y_max))

# Plot samples by color and add legend
scatter = ax.scatter(X[:, 0], X[:, 1], s=150, c=y, label=y, edgecolors="k")
ax.legend(*scatter.legend_elements(), loc="upper right", title="Classes")
ax.set_title("Samples in two-dimensional feature space")
plt.savefig("./figures/nonlinear_svm_samples.pdf", dpi=600, bbox_inches="tight")
_ = plt.show()
```

```python
def plot_training_data_with_decision_boundary(
    svc, kernel, ax=None, long_title=True, support_vectors=True, constant=1
):
    # Train the SVC
    clf = svc

    # Settings for plotting
    if ax is None:
        _, ax = plt.subplots(figsize=(10, 8))
    x_min, x_max, y_min, y_max = -1, 1, -1, 1
    ax.set(xlim=(x_min, x_max), ylim=(y_min, y_max))

    # Plot decision boundary and margins
    common_params = {"estimator": clf, "X": X, "ax": ax}
    DecisionBoundaryDisplay.from_estimator(
        **common_params,
        response_method="predict",
        plot_method="pcolormesh",
        alpha=0.1,
    )
    DecisionBoundaryDisplay.from_estimator(
        **common_params,
        response_method="decision_function",
        plot_method="contour",
        levels=[-1, 0, 1],
        colors=["k", "k", "k"],
        linestyles=["--", "-", "--"],
    )

    if support_vectors:
        # Plot bigger circles around samples that serve as support vectors
        ax.scatter(
            clf.support_vectors_[:, 0],
            clf.support_vectors_[:, 1],
            s=150,
            facecolors="none",
            edgecolors="k",
        )

    # Plot samples by color and add legend
    ax.scatter(X[:, 0], X[:, 1], c=y, s=30, edgecolors="k")
    ax.legend(*scatter.legend_elements(), loc="upper right", title="Classes")
    if long_title:
        ax.set_title(f" Decision boundaries of {kernel} kernel in SVC")
    else:
        ax.set_title(kernel)

    plt.savefig(
        f"./figures/nonlinear_svm_{kernel}_decision_boundary_{constant}.pdf",
        dpi=600,
        bbox_inches="tight",
    )

    if ax is None:
        plt.show()
```

## Linear-kernel SVM

```python
linear_results = []
C_values = [1, 100, 1000]

for C in C_values:
    lin_clf = SVC(kernel="linear", C=C)
    lin_clf.fit(X, y)

    y_hat = lin_clf.predict(X)
    zero_one_loss_value = n_obs * zero_one_loss(y, y_hat)

    linear_results.append(
        {
            "C": C,
            "Training Loss": zero_one_loss_value,
        }
    )

    plot_training_data_with_decision_boundary(lin_clf, "linear", constant=C)

print("\nLinear SVM Results Summary:")
linear_df = pd.DataFrame(linear_results)
print(linear_df)
# print(linear_df.to_latex(index=False))
```

## Gaussian / RBFkernel SVM

```python
print("\nGaussian (RBF) kernel SVM   [ = 1/(2a),   C = 1]")
rbf_results = []
a_values = [0.1, 1, 10]
# a_values = [0.1]
for a in a_values:
    gamma = 1.0 / (2.0 * a**2)  # scikit-learn parameter
    rbf_clf = SVC(kernel="rbf", C=1, gamma=gamma)
    rbf_clf.fit(X, y)

    y_hat = rbf_clf.predict(X)
    zero_one_loss_value = n_obs * zero_one_loss(y, y_hat)

    print(
        f"   a = {a:<4} ( = {gamma:0.4f}) -> training zero one loss = {zero_one_loss_value:0.3f}"
    )

    rbf_results.append(
        {
            "a": a,
            "Training Loss": zero_one_loss_value,
        }
    )

    plot_training_data_with_decision_boundary(rbf_clf, "RBF", constant=a)

print("RBF SVM Results Summary:")
rbf_df = pd.DataFrame(rbf_results)
print(rbf_df)
# print(rbf_df.to_latex(index=False))
```
